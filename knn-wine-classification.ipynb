```python
import numpy as np 
import pandas as pd 
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
wine = datasets.load_wine()
```

```python
wine
```

```python
df = pd.DataFrame(wine["data"], columns = wine["feature_names"])
```

```python
df
```

```python
df["target"] = wine["target"]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
X = df
y = X.pop("target")
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.unique()
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, val_y = train_test_split(X, y, test_size=0.2, random_state=56)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train , y_train)
```

```python
knn.score(X_val ,val_y)
```

```python
import matplotlib.pyplot as plt
k_range = range (1, 25)
scores = []
for k  in k_range:
    knn = knn = KNeighborsClassifier(k)
    knn.fit(X_train , y_train)
    scores.append(knn.score(X_val ,val_y))
plt.figure()
plt.xlabel("K count")
plt.ylabel("Model Accuracy")
plt.scatter(k_range, scores)
plt.grid()
plt.xticks([0,5,10,15,20,30])
plt.show()
```

```python
test_sizes = [0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1]
knn = KNeighborsClassifier(n_neighbors = 3)
plt.figure()
plt.xlabel("Training split %")
plt.ylabel("Accuracy")
plt.grid()
for test_size in test_sizes:
    scores = []
    
    for i in range (1, 1000):
        X_train, X_val, y_train, val_y = train_test_split(X, y, test_size=1-test_size)
        knn.fit(X_train , y_train)
        scores.append(knn.score(X_val ,val_y))
        plt.scatter(test_size, np.mean(scores))
plt.show()
```

```python
prediction = knn.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(val_y,prediction)
```
